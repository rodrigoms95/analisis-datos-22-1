```python
# Calcular y grafica autocorrelaciones.

import pandas as pd

from matplotlib import pyplot as plt
```

```python
path = "../datos/"
fname = "A.1_Wilks.csv"

# Se cargan los datos de la tabla 1 de Wilks.
df = pd.read_csv(path + fname, index_col = "Date")
```

```python
# Nombres de las columnas.
temp = ["Ithaca - Min Temp", "Ithaca - Max Temp",
    "Canandaigua - Min Temp", "Canandaigua - Max Temp"]

legend = []

for i, value in enumerate(temp):
    # Se grafica la autocorrelación hasta k = 7.
    pd.plotting.autocorrelation_plot(
        df[temp[i]], marker = "o").set_xlim([1, 7])
    # Se agrega, la simbología, "_" indica una 
    # etiqueta que no aparece.
    legend += ["_"] * 5 + [value]

plt.legend(legend)
plt.title("Autocorrelación", fontsize = 18)
```
