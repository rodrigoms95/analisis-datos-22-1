```python
import numpy as np
import pandas as pd

import xarray as xr
```

```python
path = "../datos/"
fname = "air.mon.mean.nc"

ds = xr.open_dataset(path + fname)

ds
```

```python
ds_2 = ds.sel( level = 925 )
ds_2 = ds_2.isel( lat = slice(0, 18) )
ds_2 = ds_2.isel( time = ds_2.time.dt.month == 9 )
ds_2 = ds_2.isel( time = ds_2.time
    .dt.year.isin( range(1969, 2020) ) )

ds_2
```
