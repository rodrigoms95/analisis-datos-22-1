```python
# Hace una regresi贸n no lineal.

import pandas as pd
import numpy as np

from scipy import stats

from matplotlib import pyplot as plt
```

```python
#Datos.
x = np.array([1, 2.6, 3.3, 5, 6.5, 8, 9])
y = np.array([2, 2.2, 2.5, 3.69, 6.19, 6.36, 10])

# Regresi贸n exponencial.
lin_reg = stats.linregress( x, np.log(y) )

a = lin_reg.slope
b = np.exp( lin_reg.intercept )

x_2 = np.linspace( 0, 10, 1000 )

# Y gorro.
y_reg_t = ( lin_reg.intercept +
    lin_reg.slope * x )
y_reg = b * np.exp( a * x_2 )

# Se grafican los valores transformados.
fig, ax = plt.subplots()
ax.scatter(x, np.log(y))
ax.plot(x, y_reg_t, color = "r")

ax.set_title("X - Y - Transformados",
    fontsize = 16 )
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend(["Regresi贸n\nno lineal", "Datos"])

# Se grafican los valores.
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x_2, y_reg, color = "r")

ax.set_title("X - Y",
    fontsize = 16 )
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend(["Regresi贸n\nno lineal", "Datos"])
```

```python
# Datos.
x = np.array( [1, 3, 7.34, 20.1, 40, 54.6] )
y = np.array( [6.5, 10.3, 13.5, 18.2, 21, 22.2] )

# Power regression.
lin_reg = stats.linregress( np.log(x), np.log(y) )

a = lin_reg.slope
b = np.exp( lin_reg.intercept )

x_2 = np.arange( 0, 55, 0.1 )

# Y gorro.
y_reg_t = ( lin_reg.intercept +
    lin_reg.slope * np.log(x) )
y_reg = b * x_2 ** a

# Se grafican los valores transformados.
fig, ax = plt.subplots()
ax.scatter(np.log(x), np.log(y))
ax.plot(np.log(x), y_reg_t, color = "r")

ax.set_title("X - Y - Transformados",
    fontsize = 16 )
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend(["Regresi贸n\nno lineal", "Datos"])

# Se grafican los valores.
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.plot(x_2, y_reg, color = "r")

ax.set_title("X - Y",
    fontsize = 16 )
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.legend(["Regresi贸n\nlineal", "Datos"])
```
